# Causality and VAR

The purpose of this section is to examine causal
effect estimation when the data generating process is a vector autoregressive
model.

To ground the work, let's assume that we are measuring three variables: mood, $m[k]$,
daily spending, $s[k]$ and hours exercised, $h[k]$.

We will assert an acyclic causal graph where hours exercised yesterday increases
mood which decreases spending and hours exercised today reduces spending
tomorrow.

We will also assume that each variable today is slightly correlated with its own
value yesterday.

$$
\begin{bmatrix}
h[k+1] \\
m[k+1] \\
s[k+1] \\
\end{bmatrix}
=
\begin{bmatrix}
\rho & 0 & 0 \\
\alpha & \rho & 0 \\
\beta & \gamma & \rho \\
\end{bmatrix}
\begin{bmatrix}
h[k] \\
m[k] \\
s[k] \\
\end{bmatrix} 

+
\bm{\epsilon}
$$

This implies the following DAG structure:

```julia

dot"""
digraph G {
    node [shape=plaintext]
    "hours ex" -> mood;
    mood -> spend;
    "hours ex" -> spend;
}
"""
```

Where $\bm{\epsilon} \sim \text{MVN}(\bm{0}, \sigma I)$. Note that correlated
noise would be more standard for a VAR model. Correlated noise should be
equivalent to same day correlations between variables. For compactness, let
$$
\bm{y}[k] = \begin{bmatrix}
h[k] \\
m[k] \\
s[k] \\
\end{bmatrix} 
\quad
A = \begin{bmatrix}
\rho & 0 & 0 \\
\alpha & \rho & 0 \\
\beta & \gamma & \rho \\
\end{bmatrix}
$$
so that
$$
\bm{y}[k+1] = A \bm{y}[k] + \bm{\epsilon}
$$

The research studies the effect of an intervention where each participant
exercises a fixed amount, $h_0$ per day. We can describe the intervened system
with 
$$
A_{\text{do }h_0} = \begin{bmatrix}
1 & 0 & 0 \\
\alpha & \rho & 0 \\
\beta & \gamma & \rho \\
\end{bmatrix}
\quad
\bm{\epsilon}_{\text{do }h_0} = 

\begin{bmatrix}
0 \\
\epsilon_2 \\
\epsilon_3 \\
\end{bmatrix}

\text{ and }
h_{\text{do }h_0}[0] = h_0
$$

Where $\epsilon_2$ and $\epsilon_3$ are distributed according to $\text{N}(0,
\sigma)$.


To simulate a research study, we assume that applicants are selected and are
randomly assigned to the treatment or non treatment groups. This effectively
breaks correlations between external variables and differences in applicant outcomes
can be attributed to the intervention.

For this reason it is admissible to model one group of applicants using the first VAR
and another group of applicants using the intervened model.

The goal of the study is to predict the direct effect of hours exercised on
money spent. In order to study this, the researchers instruct half the group to
exercise for one hour a day every day for $w$ weeks and give no assignment to the
other half of study participants.

At the end of the study, the researchers have participants take a survey
reporting how much money they spent in the final week along with their average
mood during the final week.

### Exercise Mood and Spending Model

```julia
"""
    mood_spend_exer_model(
        y0::Array{T, 1},
        params::Array{T, 1},
        n_steps::Int,
    ) where T

Simulates the VAR Model described above.

Args:
    * `y0` A length 3 array. The initial condition of the system.
    * `params` A length 5 array of model parameters. Should unpack 
       as follows `ρ, α, β, γ, σ = params` where the parameters correspond
       to the model defined above.
    * `n_steps` The number of timesteps to simulate

Returns:
    A 2D array with three rows and `n_steps` columns representing the
    trajectories of hours excercised, mood and money spent each day

"""
function mood_spend_exer_model(
    y0::Array{T, 1},
    params::Array{T, 1},
    nsteps::Int,
) where T

    ρ, α, β, γ, σ = params
    A = [ 
        ρ 0 0;
        α ρ 0;
        β γ ρ
    ]
    yvals = zeros(T, 3, nsteps)
    y = y0
    # Compute each step forward in time
    for i in 1:nsteps
        yvals[:, i] = y
        # Draw stochastic shocks
        ϵ = randn(3) * σ
        # Compute the next point
        y = A * y + ϵ
    end
    return yvals
end
```

## Intervened Model

```julia
"""
    mood_spend_exer_model(
        y0::Array{T, 1},
        params::Array{T, 1},
        n_steps::Int,
        h0: T,
    ) where T

Simulates the itervened VAR Model described above.

Args:
    * `y0` A length 3 array. The initial condition of the system.
    * `params` A length 5 array of model parameters. Should unpack 
       as follows `ρ, α, β, γ, σ = params` where the parameters correspond
       to the model defined above.
    * `n_steps` The number of timesteps to simulate
    * `h0` The intervention on exercise. Number of hours study participants
        exercise each day

Returns:
    A 2D array with three rows and `n_steps` columns representing the
    trajectories of, from top to bottom, hours excercised, mood and money spent
    each day. Because of the intervention, hours exercised will always be equal
    to `h0`.
"""
function mood_spend_exer_model(
    y0::Array{T, 1},
    params::Array{T, 1},
    nsteps::Int,
    h0::T
) where T

    ρ, α, β, γ, σ = params
    A = [ 
        1 0 0;
        α ρ 0;
        β γ ρ
    ]
    yvals = zeros(T, 3, nsteps)
    y = y0

    # Compute each step forward in time
    for i in 1:nsteps
        y[1] = h0
        yvals[:, i] = y
        # Draw stochastic shocks
        ϵ = randn(3) * σ
        # Compute the next point
        y = A * y + ϵ
    end
    return yvals
end
```

## Plot of The Time Series with and Without Intervention

```julia
y0 = rand(3)
params = [0.8, .2, .15, .4, .3]
nsteps = 1000

# Trajectory
Y = mood_spend_exer_model(y0, params, nsteps)

# Intervened trajectory
Y_inter = mood_spend_exer_model(y0, params, nsteps, 1.0)

plot(
    transpose(Y),
    color="green",
    label=["Control" nothing nothing]
)
plot!(
    transpose(Y_inter),
    color="orange",
    label=["Intervened" nothing nothing]
)
```

### Simulate a 35 day 50 individual study 

```julia
"""
    excercise_study(num_days::Int, num_people::Int)

Produces a (`num_people` x 2) dataframe, `df` with columns:
`["treatment", "last_week_avg_mood", "last_week_avg_spending"]`
"""
function excercise_study(;
    num_days::Int = 35,
    num_people::Int = 200,
    params::Array = [0.8, .2, .15, -0.4, .3],
    h0::Float64 = 0.25
)
    num_control = num_people ÷ 2
    num_treatment = num_people - num_control
    # Simulate control
    control_rows = mapreduce(hcat, 1:num_control) do i
        y0 = rand(3)
        Y = mood_spend_exer_model(y0, params, num_days)
        # Average of the last week
        survey = mean(Y[1:end, (end - 7):end], dims=2)
        # Mark as non-treatment
        survey[1] = 0
        survey
    end

    # Simulate treatment
    treatment_rows = mapreduce(hcat, 1:num_treatment) do i
        y0 = rand(3)
        Y_t = mood_spend_exer_model(y0, params, num_days, h0)
        # Average of the last week
        survey = mean(Y_t[1:end, (end - 7):end], dims=2)
        # Mark as treatment
        survey[1] = 1.0
        survey
    end

    results = hcat(control_rows, treatment_rows)
    df = DataFrame(
        transpose(results),
        ["treatment", "last_week_avg_mood", "last_week_avg_spending"]
    )
    return df
end
    
```

#### Total Effect of Exercise on Spending

```julia
# Bayesian linear regression.
@model function total_effect_regr(treatment, spending)
    # Set variance prior.
    σ² ~ truncated(Normal(0.0, 100); lower=0)

    # Set intercept prior.
    intercept ~ Normal(0, sqrt(3))

    # Set the priors on our coefficients.
    treatment_effect ~ Normal(0, 10)

    # Calculate all the mu terms.
    mu = intercept .+  treatment_effect .* treatment
    return spending ~ MvNormal(mu, σ² * I)
end
```

```julia
x = Array(df[!, ["treatment"]])[:, 1]
y = Array(df[!, "last_week_avg_spending"])
model = total_effect_regr(x, y)
chain = sample(model, NUTS(), 3_000; progress=false)
plot(chain)
```

We can infer the average treatment effect

```julia
# Bayesian linear regression.
@model function direct_effect_regr(treatment, mood, spending)
    # Set variance prior.
    σ² ~ truncated(Normal(0.0, 100); lower=0)

    # Set intercept prior.
    intercept ~ Normal(0, sqrt(3))

    # Set the priors on our coefficients.
    treatment_effect ~ Normal(0, 10)
    mood_effect ~ Normal(0, 10)

    # Calculate all the mu terms.
    mu = intercept .+  treatment_effect .* treatment + mood_effect .* mood
    return spending ~ MvNormal(mu, σ² * I)
end
```

```julia
treat = Array(df[!, "treatment"])
mood = Array(df[!, "last_week_avg_mood"])
spending = Array(df[!, "last_week_avg_spending"])
model = direct_effect_regr(treat, mood, spending)
chain = sample(model, NUTS(), 3_000; progress=false)
plot(chain)
```

## Summary of Simulated Study

The experiments in this notebook clairified the relationship between causal
effect estimation in randomized controled experiments and causal effect
estimation from observational data.

When you have a ramdomized controlled experiment, the analysis is simple. It
does not need to take into account time dynamics because you can average away
the dynamics or measure the system at two points and regress on those.

However, when using observational data, you don't have a clean analysis from a
regression perspective. The technique here is to model the treatment assignment
process along with the causal effect.

Another approach to this problem is to try to understand and model the structure
of the system so that the effect of interventions can be predicted.

This is a place where there is room for dynamic modeling, hence the success of
VAR, which models a single time series as the response of a system to stochastic
shocks and therefore captures something about the structure and can make
predictions about the impulse response.

This investigation highlights a way of expressing extrapolation as a formal
problem: Given a collection of one or more historic trajectories, a target
variable and two proposed interventions, predict which intervention will have a
greater positive effect on the target variable




## Best Intervention Prediction

The next section investigates the ability of a technique to use observational
data to infer which intervention will lead to a better result.

```julia
# Define Lotka-Volterra model.
function lotka_volterra(du, u, p, t)
    # Model parameters.
    α, β, γ, δ = p
    # Current state.
    x, y = u

    # Evaluate differential equations.
    du[1] = (α - β * y) * x # prey
    du[2] = (δ * x - γ) * y # predator

    return nothing
end

function intervene_lotka_volterra(du, u, p, t)
    # Model parameters.
    α, β, γ, δ = p
    # Current state.
    x, y = u

    # Evaluate differential equations.
    du[1] = 0 # prey
    du[2] = (δ * x - γ) * y # predator

    return nothing
end

# Define initial-value problem.
u0 = [5.0, 20]
p = [10.0, 3.5, 6.0, 0.5]
tspan = (0.0, 10.0)
prob = ODEProblem(lotka_volterra, u0, tspan, p)
intervene_prob = ODEProblem(intervene_lotka_volterra, u0, tspan, p)

# Plot simulation.
plot(solve(prob, Tsit5()), label=["prey" "predator"])
plot!(solve(intervene_prob, Tsit5()), label=["intervened prey" "predator response"])
```

```julia
sol = solve(prob, Tsit5())

extrapolation_prob = Dict(
    "observation" => sol.u,
    "times" => sol.t,
    "intervention_var_idx" => 1,
    "perfect_intervention" => 5.0,
    "target_var_idx" => 2,
    "labels" => Dict(
        "target_incr_relative_to_obs" => false,
        "linear" => false,
        "stationary" => true,
        "ergodic" => true,
        "chaotic" => false,
    )

)
```

```julia
function keynesian_cross(du, u, p, t)
    a, b, G = p
    I, C = u
    du[1] = I - a * C  # Investment derivative
    du[2] = b * (I - C - G)  # Consumption derivative
end

function intervened_system(df, i)
     function intervened_deriv(x...)
        df(x...)
        x[1][i] *= 0.0
     end
     return intervened_deriv
end

u0 = [10, 2]
params = [3, 2.0, 5]
tspan = (0, 20)
prob = ODEProblem(intervened_system(keynesian_cross, 1), u0, tspan, params)
plot(solve(prob, Tsit5()))
```
